```python
# set backend
lite = True

# construct an MRipl and load a model
def additive_sprinkler(no_ripls=2,lite=False):
    clear_all_engines()
    v = MRipl(no_ripls,lite=lite)
    v.assume('r','(normal 0 30)',label='r')  # RAIN
    v.assume('s','(normal 0 30)',label='s') # SPRINKLER
    v.assume('w','(lambda () (normal (+ r s) 5.))',label='w')   # WETNESS OF GRASS IS ADDITIVE IN RAIN, SPRINKLER
    v.observe('(w)','50.',label='w1')
    v.observe('(normal (w) 3.)','50.',label='w2')
    return v
```

```python
v=additive_sprinkler(no_ripls=20,lite=lite)

# create snapshot of values for RAIN, print values and plot histogram and kde
out = v.snapshot('r',plot=True)
print 'values:', out['values']
```

```python
# Customize the snapshot figure (color, annotations, figsize), save it, plot again inline

fig = out['figs'][0]
ax = fig.axes; ax0 = ax[0]
ax0.set_title('New Title for Snapshot of Variable "Rain"')
ax0.set_axis_bgcolor('gray')

ax[1].annotate("Annotation: Mean of 'r'",xy=(0,5))
ax[1].set_xlim([-60,60])

fig.set_size_inches(12,3)
fig.tight_layout()

#fig.savefig('/home/username/snapshot_rain.png')
fig.show()
```

```python
# create scatter plots of pairs of variables before and after inference

snap1=v.snapshot(['r','s'],plot=True,scatter=True)
v.infer(100)  # recall we get two noisy observations of WET GRASS = 50
snap2=v.snapshot(['r','s'],plot=True,scatter=True)

# resize figures before plotting
figs = snap1['figs']+snap2['figs']
[fig.set_size_inches(7,3.5) for fig in figs]
[ [ax.set_axis_bgcolor('.9') for ax in fig.axes] for fig in figs ]
[fig.tight_layout() for fig in figs]
plt.show()
```

```python
# MULTI mripls: two multiripls with different models of same data

# we condition the value of rain on three noisy measurements of the grass being unusually wet (=50)
# the threshold model won't favor values beyond the threshold, while the additive model will 
# favor such values (until they get too high)

def add_thresh_sprinkler(no_ripls=2,add=True):
    'either wet grass is sum of rain and sprinkler, or a simple threshold function'
   
    v = MRipl(no_ripls)
    v.assume('r','(normal 0 15)',label='r')
    v.assume('s','(normal 0 15)',label='s')
    if add:
        v.assume('w','(lambda () (normal (+ r s) 5.) )',label='w')
    else:
        v.assume('w','(lambda () (if (> (+ r s) 20) (normal 30 5) (normal 0 5) ) ) ',label='w' )
    v.observe('(w)','50.')
    v.observe('(w)','50.')
    v.observe('(w)','50.')
    return v

clear_all_engines()

## Construct one mripl for additive model and one for threshold model, conditioning on same data
v_add=add_thresh_sprinkler(no_ripls=40,add=True)
v_thresh=add_thresh_sprinkler(no_ripls=40,add=False); None
```

```python
# sample from prior and posterior for both models
snap_add1 = v_add.snapshot('r',plot=True)
v_add.infer(100)
snap_add2 = v_add.snapshot('r',plot=True)
```

```python
snap_thresh1 = v_thresh.snapshot('r',plot=True)
v_thresh.infer(100)
snap_thresh2 = v_thresh.snapshot('r',plot=True)
```
