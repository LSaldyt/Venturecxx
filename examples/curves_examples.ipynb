## Single Ripl Plots

Curves and conditional distributions for single ripls sampled from the prior of different models. Each row shows a different model. First column is the data and a sample curve (with +- 1sd noise) from the prior. Second column is samples from that curve. Third is heatmap. Following inference, we again show curves and conditional distributions. (Pivot is a quadratic that switches coefficients at a random pivot point).

We use single ripls here, but we can also map this function across an mripl and sample a small number of the curves from that mripl. 

```python
xs=list(np.random.normal(-0.5,1.4,8)) + list(np.random.normal(0.5,1.4,8) )   # FOR CRP ON X
ys = list( 5*np.sin(xs))                                    
xys=zip(xs,ys)
vs=[mk_c() for i in range(4)]  #LITE OR PUMA
[v.set_seed(i*10) for v in vs]
vs[0].execute_program(simple_quadratic_model); vs[1].execute_program(simple_fourier_model)
vs[2].execute_program(x_model_t+pivot_model); vs[3].execute_program(x_model_t+pivot_model)
vs=vs[0:3]
out=[plot_conditional(v,xr=(-3.5,3.5),data=xys,no_reps=20,no_xs=30) for v in vs];
```

```python
observe_infer(vs,xys,1000)
out=[plot_conditional(v,xr=(-3.5,3.5),data=xys,no_reps=20,no_xs=30) for v in vs];
```

## MRipl Plots

Now we show samples of the conditional of y on X=x from multiple ripls (via an mripl). For the simple fourier model, samples from the prior average out to a straight line. 

```python
xs=list(np.random.normal(-0.5,1.4,6)) + list(np.random.normal(0.5,1.4,6) )   # FOR CRP ON X
ys = list( 2*np.sin(xs))                                    
xys=zip(xs,ys)
v=MRipl(6,lite=False); v.execute_program(simple_fourier_model);out=posterior_conditional(v,xr=(-3.5,3.5),data=xys)
observe_infer([v],xys,10)
```
