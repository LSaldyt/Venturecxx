````python
from venture.venturemagics.ip_parallel import *
from venture.unit import Analytics

def make_random_model(my_args=None):
    
    args = dict(
        prior_switch_point = '(gamma 1 .01)',
        prior_lam_linear = '(gamma 1 .3)',
        prior_post_switch = '(gamma 1 .01)',
        prior_random_rate = '(gamma 1 .01)',
        dist_random_event = '(flip .01)' )
    
    if my_args: args.update(my_args) 
    

    random_model='''
    [assume lam_start (scope_include (quote hypers) 0 
                        {prior_lam_linear} )]
    [assume lam_coefficient (scope_include (quote hypers) 1 
                        {prior_lam_linear} )]

    [assume switch_point (scope_include (quote switch) 0 
                           {prior_switch_point})]
    [assume lam_post_switch (scope_include (quote hypers) 2 
                           {prior_post_switch})]
    [assume random_rate (scope_include (quote hypers) 3 
                           {prior_random_rate})]

    [assume random_event (lambda () {dist_random_event})]

    [assume rate (mem (lambda (t)
                        (scope_include (quote rate_choice) t
                          (if (random_event)
                            random_rate
                            (regular_rate t)))))]

    [assume regular_rate (lambda (t) 
                          (if (< t switch_point) 
                            (+ lam_start (* t lam_coefficient))
                            lam_post_switch ))]

    [assume g2b (lambda (t) (poisson (rate t))) ]
    '''.format( prior_switch_point = args['prior_switch_point'], prior_lam_linear = args['prior_lam_linear'],
                prior_post_switch = args['prior_post_switch'], prior_random_rate = args['prior_random_rate'],
                dist_random_event = args['dist_random_event'])
    return random_model
````

````python
def test_random_model():
    v=mk_p_ripl()
    num_draws = 150
    args = dict(
        dist_random_event = '(flip .9999)',
        prior_random_rate = 100,)
    v.execute_program( make_random_model(args) )
    
    assert int( v.sample('random_rate') ) == int( args['prior_random_rate'] )
    g2bs =[v.predict('(g2b {})'.format(i)) for i in range(num_draws)]
    rates = map(int, [v.predict('(rate {})'.format(i)) for i in range(num_draws)] )
    for r in rates:
        assert r == args['prior_random_rate']
    assert np.abs( (np.mean(g2bs) - args['prior_random_rate']) ) < 5
    assert np.abs( (np.var(g2bs) - args['prior_random_rate']) ) < 30
    
    
    args = dict(
                 dist_random_event = '(flip .0001)',
                 prior_lam_linear = 1)
    v.clear()
    v.execute_program( make_random_model(args) )
    v.print_directives()
    assert int( v.sample('lam_start')) == args['prior_lam_linear']
    assert int( v.sample('lam_coefficient')) == args['prior_lam_linear']
    rates = map(int, [v.predict('(rate {})'.format(i)) for i in range(num_draws)] )
    for i,r in enumerate( rates ):
        assert int(r) == i+1
    
test_random_model()
````

````python
from venture.venturemagics.ip_parallel import *
from venture.unit import Analytics

# not yet added to model, but can replace 'lam' in m1 with 
# this random rate


random_rate = '''
[assume random_event (lambda () (flip .01))]
[assume random_rate (scope_include (quote hypers) 3 (gamma 1 .02))]
[assume rate (mem (lambda (t)
                    (scope_include (quote rate_choice) t
                      (if (random_event)
                        (random_rate)
                        (regular_rate t)))))]
                            
[assume regular_rate (lambda (t) 
                      (if (< t switch_point) 
                        (+ lam_start (* t lam_coefficient))
                        lam_post_switch ))]'''

# current model
# lam is the poisson lambda that varies with time

switch_prior = '(gamma 1 .01)'
lam_hypers = '(gamma 1 .3)'

m1='''
[assume lam_start (scope_include (quote hypers) 0 %s )]
[assume lam_coefficient (scope_include (quote hypers) 1 %s )]

[assume switch_point (scope_include (quote switch) 0 %s)]
[assume lam_post_switch (scope_include (quote hypers) 2 (gamma 1 .01))]

[assume lam (lambda (t) (if (< t switch_point) 
                          (+ lam_start (* t lam_coefficient))
                          lam_post_switch ))]
[assume g2b (lambda (t) (poisson (lam t))) ]
'''% (lam_hypers,lam_hypers,switch_prior)

# basic model
m='''
[assume lam_start (scope_include (quote hypers) 0 (gamma 3 1) )]
[assume lam_coefficient (scope_include (quote hypers) 1 (gamma 3 1) )]

[assume lam (lambda (t) (+ lam_start (* t lam_coefficient)))]
[assume g2b (lambda (t) (poisson (lam t))) ]

'''
v=MRipl(2,local_mode=True)
v.execute_program(m1)
lam_start = 3; lam_coefficient = 2
num_timesteps = 200
switch_point = int(.5*num_timesteps)
lam_post_switch = lam_start + (switch_point *lam_coefficient)

observes = []; queries = []
poi = lambda lam: np.random.poisson(lam)

for i in range(0,num_timesteps,3):
    if i < switch_point:
        [observes.append( ( '(g2b %i)'%i, poi(lam_start + (i*lam_coefficient)) )) for _ in range(1)]
    else:
        [observes.append( ( '(g2b %i)'%i, lam_post_switch ) ) for _ in range(1)]
    queries.append( '(lam %i)'%i)
[v.observe(*obs) for obs in observes];
ana = Analytics(v);

def plot_hypers(h):
    map( h.quickPlot, ('lam_start','lam_coefficient','switch_point','lam_post_switch') )
````
