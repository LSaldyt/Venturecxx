## Inferring the Mean and Variance of a Gaussian Distribution

### 1. Specifying and plotting the prior



These are the heights of a group of a 18-year-old men. We'd like to make an inference about the population based on this data.

````python
male18=[179,195.1,183.7,178.7,171.5,181.8,172.5,
        174.6,190.4,173.8,172.6,185.2,178.4,
        177.6,183.5,178.1,177,172.9,188.4,169.4,
        180.2,189,182.4,185.8,180.7,178.7]
data=male18
fig,ax=plt.subplots(figsize=(4,2))
ax.hist(data, alpha=.6, color='orange')
ax.set_title('Histogram of Heights for 18 year-old Males');
````

We model the population distribution as a Gaussian with unknown mean `mu` and standard deviation `sigma`. We place broad priors on each of the variables. We then define `x` as a stochastic procedure that samples a height i.i.d. given `mu` and `sigma`.

````python
v=MRipl(80)
````

````python
%%venture v
[assume mu (normal 0 60) ]
[assume sigma (gamma 2 .2) ]
[assume x (lambda () (normal mu sigma) ) ]
````

We can visualize our prior on `mu` and `sigma`, as well as their covariance, via the 'scatter' option for `snapshot`.

````python
v.snapshot(('mu','sigma'), plot=True, scatter=True);
````

Each pair (`mu`, `sigma`) produces a distinct Gaussian population distribution generated from sampling `x`. Using the MRipl method `sample_populations`, we can plot some of these populations (sampled from the model prior).

The form is:
```python
<mripl>.sample_populations(<stochastic_expression>, <number_populations>, <population_size>)
```

````python
out=v.sample_populations('(x)', 5, 100) 
````

````python
out=v.sample_populations('(x)', 4, 100)
````

````python
squared_x = v.sample_populations('(pow (x) 2)', 4, 100)
````

If we sample a population from each RIPL in the MRipl and collapse all individuals into one big population, we get an approximation of the marginal prior on heights.

````python
outs=v.snapshot('(x)',plot=True,repeat=60)
````

### 2. Conditioning the model on data

We condition the model using the `observe` directive. As with the 'Tricky Coin' example, we can either condition _incrementally_, interleaving `observe` and `infer` directives, or we can condition on all the data at once. Here we exhibit incremental conditioning.

````python
prior_mu = v.snapshot('mu')

v.observe('(x)', data[0] )
print 'Observed data: ', data[:1]

v.infer(500)
posterior1_mu=v.snapshot('mu')
fig = v.compare_snapshots( (prior_mu,posterior1_mu) ) 
````

````python
v.observe('(x)',data[1] )
print 'Observed data: ', data[:2]

v.infer(500)
posterior2_mu=v.snapshot('mu')

# NOTE: we are comparing the posterior given two data-points to the original prior:
# i.e. the prior before conditioning on any data
fig = v.compare_snapshots( (prior_mu,posterior2_mu) ) 
````

````python
v.observe('(x)', data[2] )
v.observe('(x)', data[3] )
print 'Observed data: ', data[:4]

v.infer(500)
posterior3_mu=v.snapshot('mu')
fig = v.compare_snapshots(prior_mu, posterior1_mu, posterior2_mu, posterior3_mu)
````

## Making inference robust to outliers

````python
v=MRipl(40)
````

````python
%%venture v
[assume mu (normal 0 60) ]
[assume sigma (gamma 2 .2) ]
[assume x (lambda () (normal mu sigma))]
````

````python
[v.observe('(x)',str(data[i])) for i in range(4) ]
v.infer(600);
posterior_before_outlier=v.snapshot(exp_list='mu',plot=True);
````

````python
v.observe('(x)',str(325))
v.infer(600)
posterior_after_outlier=v.snapshot(exp_list='mu',plot=True,plot_past_values=[posterior_before_outlier])
````

````python
vr = MRipl(20)
````

````python
%%venture vr
[assume mu (normal 0 60) ]
[assume sigma (gamma 2 .2) ]
[assume x (lambda () (if (flip .96) (normal mu sigma) (normal mu (* 10 sigma) ) ) ) ]
````

````python
# bugs with vr
[v.observe('(x)',str(data[i])) for i in range(4) ]
v.infer(600);
posterior_before_outlier=v.snapshot(exp_list='mu',plot=False);
v.observe('(x)',str(325))
v.infer(600)
posterior_after_outlier=v.snapshot(exp_list='mu',plot=True,plot_past_values=[posterior_before_outlier])
````
