# Introduction to MRipl

### What is an MRipl?

<ol><li>An MRipl makes it simple to map Venture ripl directives across a set of ripls. 
<li>Basic syntax is the same as for a ripl:<code> mripl.assume('x','(beta 1 1)')</code> will map this ASSUME directive across all ripls in the mripl.
<li>This allows inference with multiple parallel chains. Visualize the chains with <code>mripl.snapshot</code> .
<li>Map custom procedures across all ripls in the mripl with <code>mripl.map_proc</code>.
<li>Ripls can either be <b>remote</b> (via IPython Parallel) or <b>local</b> (i.e. simply a list of ripls in the local namespace). 

### 1. Getting Started

### 2. Create an MRipl, Basic Features

We begin with a simple example showing basic MRipl features.

```python
# first arg to constructor is number of ripls in the MRipl
v=MRipl(10)  # typically number of ripls will be multiple of #engines (above)

print 'Number of ripls:',v.no_ripls
print 'seeds:', v.seeds
```

```python
# there are methods for all basic ripl directives/methods
assumeValues = v.assume('p','(beta 1 1)') # we get output from each ripl
print assumeValues[:4]
```

```python
[v.observe('(flip p)',True) for _ in range(3)]
v.infer(20)
print 'Some values of *p* for ripls after inference:', v.report(1)[:4]
```

```python
# display a set of directives
display_directives(v,'observe'); 
```

```python
# collect and plot values of any expression using snapshot
out = v.snapshot(exp_list=('p',),plot=True)
```

```python
# map a custom function across ripls with map_proc

def flip_plot(ripl): # function has to take ripl as first argument
    flip_coin_20 = [ripl.sample('(flip p)') for _ in range(20)]
    fig,ax = plt.subplots(figsize=(2,2))  # Matplotlib is already imported on remote engines
    ax.hist(flip_coin_20,color='m')
    
v.map_proc(2,flip_plot);  # self.map_proc( no_ripls, proc, *proc_args, **proc_kwargs)
```

```python
# map across *all* ripls 
v.map_proc('all', lambda ripl: ripl.sample('(and (flip p) (flip p))'))
```

### 3. MRipl Attributes: Local vs. Remote modes, Seeds, Backends

Three ways to enter a set of directives:
<ol><li>Venture cell magic: <code>%%venture  mripl_name  [number_ripls_printed]</code>

```python
%%venture  v  3
[assume w0 (normal 0 3) ]
[assume w1 (normal 0 1) ]
[assume w2 (normal 0 .3) ]
[assume x_d (lambda () (normal 0 5))]

[assume f (lambda (x) (+ w0 (* w1 x) (* w2 (* x x)) ) ) ]
[assume y_x (lambda (x) (normal (f x) noise) ) ]
```

```python
# 2. Using a string
quadratic_model='''
[assume w0 (normal 0 3) ]
[assume w1 (normal 0 1) ]
[assume w2 (normal 0 .3) ]
[assume x_d (lambda () (normal 0 5))]
[assume noise (gamma 2 1) ]
[assume f (lambda (x) (+ w0 (* w1 x) (* w2 (* x x)) ) ) ]
[assume y_x (lambda (x) (normal (f x) noise) ) ]'''
v=MRipl(4)
v.execute_program(quadratic_model)

# 3. Using Python method notation (not show in full)
v=MRipl(4)
v.assume('w0','(normal 0 3)')
# ...
```

```python
# snapshot a pair of variables
v.snapshot(exp_list=['w0','w1'],scatter=True);
```

```python
# sample populations: look at the predictive distribution on x=5 for some of our ripls
v.snapshot('(y_x 5)',sample_populations=(4,20),plot=True);
```

```python
# repeat: look at overall predictive distribution by integrating across ripls
v.snapshot('(y_x 5)',repeat=20,plot=True);
```

### Mapping Functions Across the Ripls

```python
# write a function for plotting curves given a ripl
def my_plot_conditional(ripl,x_range=(-3,3),number_xs=20):
    xr = np.linspace(x_range[0],x_range[1],number_xs)
    f_xr = np.array([ripl.sample('(f %f)' %x) for x in xr])
  
    fig,ax = plt.subplots(figsize=(2,2))
    ax.plot(xr,f_xr)

# test function on the local ripl
v=MRipl(4)
v.execute_program(quadratic_model)
my_plot_conditional(v.local_ripls[0],x_range=(-3,3),number_xs=20)
```

```python
# map the function across the remote ripls, specifying optional arguments
out = mr_map_proc(v,2,my_plot_conditional,x_range=(-4,4),number_xs=20)

# note that we limit the mapping to only two ripls
```

```python
# mr_map_proc maps the same function and args across ripls
# mr_map_array allows for different args

def my_plot_conditional_color(ripl,color):
    xr = np.linspace(-4,4,30)
    f_xr = np.array([ripl.sample('(f %f)' %x) for x in xr])
    fig,ax = plt.subplots(figsize=(2,2))
    ax.plot(xr,f_xr,color='black')
    ax.fill_between(xr,f_xr,min(f_xr),alpha=.4,color=color)

out = mr_map_array(v,my_plot_conditional_color,['y','m'])
```

```python
## mr_map_array allows us to vary directives across the ripls
## (all other methods on an mripl leave us with parallel chains)

# example: conditioning on different data for each engine

def regression_observe(ripl,data):
    for (x,y) in data:
        ripl.observe('(x_d)', '%f' % x )
        ripl.observe('(y_x %f)' % x , '%f' % y )
        
data1 = [(-3,-3),(-2, -2.3), (.05, 0), (2,2), (3,3)]
data2 = [(-3,9.2),(-2, 4), (0.3,0.1),(2.1,4),(3,9)]

fig,ax = plt.subplots()
ax.scatter(*zip(*data1),c='red',label='Data for Engine 0',lw=0)
ax.scatter(*zip(*data2),label='Data for Engine 1',lw=0)
ax.legend(loc='lower right')

v=MRipl(4)
v.execute_program(quadratic_model)
out = mr_map_array(v,regression_observe,[[data1],[data1],[data2],[data2] ])
```

```python
# the function 'display_directives' will print the directives
# of a ripl or MRipl

# across each engine, the ASSUMES are identical
mr_map_proc(v,2,display_directives,'assume')
```

```python
# but the OBSERVES vary across the engines
mr_map_proc(v,2,display_directives,'observe')
```
