### Linear Regression, no AAA --> Big scaffold

```python
%%venture v
[clear]
[assume w0 (scope_include (quote global_var) 0 (normal 0 3)) ]
[assume w1 (scope_include (quote global_var) 0 (normal 0 1)) ]
[assume x_d (lambda () (normal 0 5))]
[assume noise (scope_include (quote global_var) 0 (gamma 2 1)) ]
[assume f (lambda (x) (+ w0 (* w1 x) ) ) ]
[assume y_x (lambda (x) (if (flip .96) (normal (f x) noise) (normal (f x) (* 10 noise) ) ) ) ]
[assume model_name (quote linear_reg_robust_to_outlier)]
```

```python
xs=list(np.random.normal(3,.7,3)) + list(np.random.normal(-1,1.4,3)) # generate x values
f = lambda x:-2*x
y_x = lambda x: np.random.normal(f(x),.6)
ys = [y_x(x) for x in xs]   # compute y values from x,f and Normal noise                                 
data = zip(xs,ys)

for i,(x,y) in enumerate(data):
    v.observe('(x_d)', '%f' % x ) 
    v.observe('(y_x %f)' % x , '%f' % y )
v.predict('(y_x 0)')
```

```python
v.infer({'kernel': 'draw_scaffold', 'scope': 'global_var', 'block' : 0, 'transitions': 1})
```
