## VentureUnit

```python
from venture import shortcuts
from venture.unit import VentureUnit
class BinaryMRF(VentureUnit):
    def makeAssumes(self):
        D = self.parameters['nodes']
        psi = self.parameters['psi']
        for d in range(D):
            self.assume('x_%d' % d, '(scope_include (quote global_var) %d (bernoulli 0.5))' % d)

        n = 0
        for d1 in range(D):
            str_x1 = 'x_%d' % d1
            for d2 in range(d1+1, D):
                str_x2 = 'x_%d' % d2

                # Potential with (X1, X2)
                str_y = '(bernoulli (+'
                for x1 in [True, False]:
                    for x2 in [True, False]:
                        # Value of (X1=x1, X2=x2)
                        str_y += ' (* %f (and %s %s))' % (psi[n],
                                                          str_x1 if x1 else '(not %s)' % str_x1, 
                                                          str_x2 if x2 else '(not %s)' % str_x2)
                        n += 1
                str_y += '))'
                self.assume('y_%d_%d' % (d1,d2), str_y)
                self.observe('y_%d_%d' % (d1,d2), True)
        return


    def makeObserves(self):
        D = self.parameters['nodes']
        for d1 in range(D):
            for d2 in range(d1+1, D):
                self.observe('y_%d_%d' % (d1,d2), True)
        return


# ripl = shortcuts.make_church_prime_ripl()
ripl = shortcuts.make_lite_church_prime_ripl()

D = 3
psi = np.exp(np.random.normal(0, np.sqrt(0.02), D * (D-1) / 2 * 2**2))

parameters = {'nodes' : D, 'psi' : psi}

model = BinaryMRF(ripl, parameters)
#history = model.runConditionedFromPrior(50, verbose=True)
#history = model.runFromJoint(50, verbose=True)
#history = model.sampleFromJoint(20, verbose=True)
# sample_hist, infer_hist, klHistory = model.computeJointKL(20, 10, verbose=True)
# klHistory.plot(fmt='png',directory='klHistory')
history,_ = model.runFromConditional(100, verbose=True)
history.plot(fmt='pdf',directory='figure_BinaryMRF_runFromConditional/')

#parameters = {'topics' : [4, 8], 'vocab' : 10, 'documents' : [8, 12], 'words_per_document' : [10, 100]}
#run_count = 0
#def runner(params):
#    print "Running setting " + str(run_count) + " of 8"
#    print params
#    return LDA(ripl, params).computeJointKL(20, 20, verbose=True)
#histories = productMap(parameters, runner)
#
#plotAsymptotics(parameters, histories, 'sweep_time', fmt='png', aggregate=True)
```

## Potential of three variables

```python
D = 4
psi = np.exp(np.random.normal(0, np.sqrt(0.02), D * (D-1) * (D-2) / 6 * 2**3))
```

```python
v.clear()
for d in range(D):
    v.assume('x_%d' % d, '(scope_include (quote global_var) %d (bernoulli 0.5))' % d)

n = 0
for d1 in range(D):
    str_x1 = 'x_%d' % d1
    for d2 in range(d1+1, D):
        str_x2 = 'x_%d' % d2
        for d3 in range(d2+1, D):
            str_x3 = 'x_%d' % d3
            
            # Potential with (X1, X2, X3)
            str_y = '(bernoulli (+'
            for x1 in [True, False]:
                for x2 in [True, False]:
                    for x3 in [True, False]:
                        # Value of (X1=x1, X2=x2, X3=x3)
                        str_y += ' (* %f (and %s (and %s %s)))' % (psi[n],
                                                                   str_x1 if x1 else '(not %s)' % str_x1, 
                                                                   str_x2 if x2 else '(not %s)' % str_x2,
                                                                   str_x3 if x3 else '(not %s)' % str_x3)
                        n += 1
            str_y += '))'
            v.assume('y_%d_%d_%d' % (d1,d2,d3), str_y)
            v.observe('y_%d_%d_%d' % (d1,d2,d3), True)
```

```python
plt.figure(figsize=(40,40))
v.infer({'kernel': 'draw_scaffold', 'scope': 'global_var', 'block' : 0, 'transitions': 1})
```
