## Test example for Exercise 1. Learn how to add a new inference operator.

Modified from gauss_reg_demo.

```python
from venture.venturemagics.ip_parallel import *
```

```python
male18=[179,195.1,183.7,178.7,171.5,181.8,172.5,
        174.6,190.4,173.8,172.6,185.2,178.4,
        177.6,183.5,178.1,177,172.9,188.4,169.4,
        180.2,189,182.4,185.8,180.7,178.7]
data=male18
fig,ax=subplots(figsize=(4,2))
ax.hist(data,color='orange'); ax.set_title('Heights of 18 y/o Males');
```

```python
# Run infer with the old kernel.
v=MRipl(4,backend='lite',local_mode=True)
```

```python
%%venture v
[assume mu (normal 0 60) ]
[assume sigma (gamma 2 .2) ]
[assume x (lambda () (normal mu sigma) ) ]
```

```python
v.observe('(x)',str(data[0]) )
print 'Observed data: ', data[:1]
v.infer({'transitions':200, 'kernel':'mh'})
posterior1_mu=v.snapshot(exp_list='mu')
```

```python
[v.observe('(x)',str(x) ) for x in data[1:]]
print 'Observed data: ', data
v.infer({'transitions':200, 'kernel':'mh'})
posteriorall_mu=v.snapshot(exp_list='mu')
fig = v.snapshot(plot_past_values=[posterior1_mu,posteriorall_mu])
```

```python
# Run infer with the new kernel.
v=MRipl(4,backend='lite',local_mode=True)
```

```python
%%venture v
[assume mu (normal 0 60) ]
[assume sigma (gamma 2 .2) ]
[assume x (lambda () (normal mu sigma) ) ]
```

```python
v.observe('(x)',str(data[0]) )
print 'Observed data: ', data[:1]
v.infer({'transitions':200, 'kernel':'subsampled_mh'})
posterior1_mu=v.snapshot(exp_list='mu')


[v.observe('(x)',str(x) ) for x in data[1:]]
print 'Observed data: ', data
v.infer({'transitions':200, 'kernel':'subsampled_mh'})
posteriorall_mu=v.snapshot(exp_list='mu')
fig = v.snapshot(plot_past_values=[posterior1_mu,posteriorall_mu])
```
