# Scaffold for Bayesian Logistic Regression

```python
from venture.venturemagics.ip_parallel import *
import ipdb
make_ripl = make_lite_church_prime_ripl
```

```python
v=make_ripl()
```

```python
D = 3
N = 5
```

```python
v.clear()
for d in range(D + 1):
    v.assume('w_%d' % d, '(scope_include (quote global_var) 0 (normal 0 (sqrt 0.1)))')

str_x = ' '.join(['x_%d' % d for d in range(1, D + 1)])
str_f = '+ w_0 ' + ' '.join(['(* w_%d x_%d)' % (d, d) for d in range(1, D + 1)])
v.assume('f', '(scope_include (quote f) 0 (lambda (%s) (%s)))' % (str_x, str_f))
v.assume('s', '(scope_include (quote s) 0 (lambda () (bernoulli 0.5)))')
v.assume('y_x', '(scope_include (quote y_x) 0 (lambda (%s) (if (bernoulli 0.8) (bernoulli (/ 1 (+ 1 (exp (- 0 (f %s)))))) (s))))' % (str_x, str_x));
```

```python
w = np.random.normal(0, np.sqrt(0.1), D + 1).tolist()
xs = np.random.normal(0, 1, [N, D])
ys = np.random.random(N) < (xs.dot(w[1:]) + w[0])
xs = xs.tolist()
ys = ys.tolist()
```

```python
for n in xrange(N):
    str_x = ' '.join(['%f' % x for x in xs[n]])
    v.observe('(scope_include (quote to_subsample) %d (y_x %s))' % (n, str_x), ys[n])
```

```python
plt.figure(figsize=(30,30))
v.infer({'kernel': 'draw_scaffold', 'scope': 'default', 'block' : 'all', 'transitions': 1})
```

```python
plt.figure(figsize=(25,25))
v.infer({'kernel': 'draw_subsampled_scaffold', 'scope': 'global_var', 'block' : 'all', 'transitions': 1, 'scope_to_subsample': 'to_subsample'})
```
